## Loading Libraries

```python
import pandas as pd
import numpy as np
```

## Loading Data

```python
# Customer Data

cust_df = pd.read_excel('./Data/Customer_Data.xlsx')
cust_df.head()
```

```python
# JTD Data

jtd_df = pd.read_csv('./Data/JTD.csv')
jtd_df.head()
```

```python
# Plant Data

plant_df = pd.read_excel('./Data/PlantDataNew.xlsx')
plant_df.head()
```

```python
# Invoice Data

invoice_df = pd.read_csv('./Data/Final_invoice.csv')
invoice_df.head()
```

```python
# Plants which are present in Invoice data but not in Plant master data

plant_inv=[]
for plant in invoice_df['Plant'].unique():
    if plant not in plant_df['Plant'].unique():
        plant_inv.append(plant)
        
print(plant_inv)
```

```python
# Distinct value of City for above Plants

for plant_1 in plant_inv:
    print(invoice_df['Plant Name1'][invoice_df['Plant'] == plant_1].unique())
```

```python
# Loading City and state JSON file

import json 
  
# Opening JSON file 
with open('./Data/City_list.json') as json_file: 
    city_name = json.load(json_file) 
  
    # Print the type of data variable 
    print("Type:", type(city_name))
  
    # Print the data of dictionary 
    print("\nCity Name:", city_name['City'])
    print("\nCity Name:", city_name['State'])
```

```python
# Replace Plant Name1 value with City

for key, value in city_name['City'].items():
    invoice_df.loc[invoice_df['Plant Name1']==key, 'Plant Name1'] = value
```

## Merging DataSet into Master

```python
# Invoice data as master

master = invoice_df.drop(columns=['Unnamed: 0','Amt Rcvd From Custom','Amt Rcvd From Ins Co','Area / Locality','CGST(14%)','CGST(2.5%)',
                              'CGST(6%)','CGST(9%)','Cash /Cashless Type','Claim No.','Cust Type','District','Expiry Date','Gate Pass Date',
                              'Gate Pass Time','IGST(12%)','IGST(18%)','IGST(28%)','IGST(5%)','Insurance Company','ODN No.','Policy no.',
                              'Outstanding Amt','Pin code','SGST/UGST(14%)','SGST/UGST(2.5%)','SGST/UGST(6%)','Print Status',
                              'SGST/UGST(9%)','Service Advisor Name','Total CGST','Total GST','Total IGST','Total SGST/UGST','Recovrbl Exp',
                              'TDS amount','Technician Name','Total Amt Wtd Tax.','Invoice No','Regn No','User ID'])

master.head(3)
```

```python
master.shape
```

```python
## Join plant details to invoice
master = pd.merge(left=master,
                   right=plant_df[['Plant','City','State','Sales organization']],
                   left_on='Plant',
                   right_on='Plant',
                   how='left')
```

```python
master.shape
```

```python
# Replace null value of city

for plant_1 in plant_inv:
    master.loc[master['Plant']==plant_1, 'City'] = invoice_df['Plant Name1'][invoice_df['Plant'] == plant_1]
```

```python
master.loc[master['City']=='HINJAWADI – PUNE', 'City'] = 'Pune'
```

```python
# Replace blank state value with State

for key, value in city_name['State'].items():
    master.loc[master['City']==key, 'State'] = value
```

```python
# Replace null values of sales organization

master.loc[((master['Sales organization'].isnull()==True) & (master['State']=='Punjab')), 'Sales organization'] = 'MFCS'
master.loc[((master['Sales organization'].isnull()==True) & (master['State']=='Tamil Nadu')), 'Sales organization'] = 'MFCS'
master.loc[master['Sales organization'].isnull()==True, 'Sales organization'] = 'MFCD'
```

```python
# Drop zero value of Total value column
master.drop(master[master['Total Value']==0].index, inplace=True)
```

```python
# Merge Make and model column to create Car Model column
master['Car Model'] = master['Make'] + ' ' + master['Model'].replace(np.nan, '')
```

```python
# Join customer details to master

master = pd.merge(left=master,
                   right=cust_df['Customer No.'],
                   left_on='Customer No.',
                   right_on='Customer No.',
                   how='left')
```

```python
master.shape
```

```python
# Join job details to master
master = pd.merge(left=master,
                   right=jtd_df[['DBM Order','Order Item','Item Category','Order Quantity','Target quantity UoM','Net value']],
                   left_on='Job Card No',
                   right_on='DBM Order',
                   how='left')
```

```python
master.shape
```

## Replace Null values

```python
# Replace null values of Total value

master.loc[master['Total Value'].isnull()==True, 'Total Value'] = master['OSL Total'] + master['Parts Total'] + master['Misc Total'] + master['Labour Total']
```

```python
# Drop null values of Target quantity UoM
master.drop(master[master['Target quantity UoM'].isnull()==True].index, inplace=True)
```

```python
# Drop columns
master.drop(columns=['CITY', 'Make', 'Model'], axis=1, inplace=True)
```

```python
# Take Citywise customer count as a column

city_cust = pd.DataFrame(master.groupby(['City'])['Customer No.'].nunique())
city_cust.reset_index(inplace=True)
city_cust.columns = ['City', 'Cust Count']
```

```python
# Take Statewise plant count as a column

state_plant = pd.DataFrame(master.groupby(['State'])['Plant'].nunique())
state_plant.reset_index(inplace=True)
state_plant.columns = ['State', 'Plant Count']
```

```python
# Join city_cust data to master

master = pd.merge(left=master,
                   right=city_cust[['City', 'Cust Count']],
                   left_on='City',
                   right_on='City',
                   how='left')
```

```python
# Join city_cust data to master

master = pd.merge(left=master,
                   right=state_plant[['State', 'Plant Count']],
                   left_on='State',
                   right_on='State',
                   how='left')
```

```python
master.shape
```

## Calculate Service Hours

```python
# Convert type of columns into DateTime

import datetime as dt

master['Invoice Date'] = master['Invoice Date'].apply(lambda r : dt.datetime.strptime(r, '%Y-%m-%d').date())
master['Invoice Time'] = master['Invoice Time'].apply(lambda r : dt.datetime.strptime(r, '%H:%M:%S').time())

master['JobCard Date'] = master['JobCard Date'].apply(lambda r : dt.datetime.strptime(r, '%Y-%m-%d'))
master['JobCard Time'] = master['JobCard Time'].apply(lambda r : dt.datetime.strptime(r, '%H:%M:%S').time())
```

```python
# Combine Date and Time

master['Invoice Date'] = master.apply(lambda r : pd.datetime.combine(r['Invoice Date'], r['Invoice Time']),1)
master['JobCard Date'] = master.apply(lambda r : pd.datetime.combine(r['JobCard Date'], r['JobCard Time']),1)
```

```python
# Calculate Service time

def cal_hours(startDate, endDate):
    difference = startDate - endDate
    hours = divmod(difference.total_seconds(), 3600)
    return hours[0]
```

```python
# Add a column with time difference

master['Service Hours'] = master.apply(lambda r : cal_hours(r['Invoice Date'], r['JobCard Date']), axis=1)
```

## Drop Columns

```python
# Drop columns

master.drop(columns=['Customer No.','Invoice Date','Invoice Time','Job Card No','JobCard Date','JobCard Time','Misc Total','OSL Total','Plant','Plant Name1','DBM Order'], axis=1, inplace=True)
```
