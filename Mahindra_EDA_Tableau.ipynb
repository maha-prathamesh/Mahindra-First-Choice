```python
# Calculate Longitude and Latitude
import pgeocode

geolocator = pgeocode.Nominatim('in')

def cal_Longitude(postal_code):
    try:
        location = geolocator.query_postal_code(postal_code)
        return location.longitude
    except:
        return 0.0

def cal_Latitude(postal_code):
    try:
        location = geolocator.query_postal_code(postal_code)
        return location.latitude
    except:
        return 0.0
```

```python
import json 
  
# Opening JSON file 
with open('./Data/ZipCode_Longi_Lati.json') as json_file: 
    zip_data = json.load(json_file) 
  
    # Print the type of data variable 
    print("Type:", type(zip_data))
  
    # Print the data of dictionary 
    print("\nLongitude:", zip_data['Longitude']) 
    print("\nLatitude:", zip_data['Latitude'])
```

```python
# Replace NaN value of Longitude
for key, value in zip_data['Longitude'].items():
    plant_data.loc[plant_data['Postal Code']==int(key), 'Longitude'] = value
    print(key,value)
```

```python
# Replace NaN value of Latitude
for key, value in zip_data['Latitude'].items():
    plant_data.loc[plant_data['Postal Code']==int(key), 'Latitude'] = value
    print(key,value)
```

```python
import json
# Opening JSON file 
with open('./Data/Customer_Metadata.json') as json_file: 
    metadata = json.load(json_file) 
  
    # Print the type of data variable 
    print("Type:", type(metadata))
  
    # Print the data of dictionary 
    print("\nPartner Type:", metadata['Partner Type']) 
    print("\nData Origin:", metadata['Data Origin'])
    print("\nTitle:", metadata['Title'])
    print("\nMarital Status:", metadata['Marital Status'])
    print("\nOccupation:", metadata['Occupation'])
```

```python
# Replace Code with desc
columns = ['Partner Type', 'Title', 'Marital Status', 'Occupation']
for col in columns:    
    for key, value in metadata[col].items():
        cust_data.loc[cust_data[col]==int(key), col] = value
```

```python
for key, value in metadata['Data Origin'].items():
        cust_data.loc[cust_data['Data Origin']==key, 'Data Origin'] = value
```

```python
cust_data['Customer No.'] = cust_data['Customer No.'].apply(str)
```
